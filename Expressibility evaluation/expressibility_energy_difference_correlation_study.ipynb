```python
%matplotlib inline
# Importing standard Qiskit libraries and configuring account
from qiskit.tools.jupyter import *
from qiskit.visualization import *
import pylab
import numpy as np
import pandas as pd
```

Noise-free expressibility correlation study 

```python
expr_data = pd.read_csv('expr_energy_data.csv', header=0, sep=';')
expr_data_df = pd.DataFrame(expr_data)
expr_data_df.head(12)
```

```python
Reference_val = -1.137306035753
expr_data_df['Ediff'] =  expr_data_df['Energy'] - Reference_val
expr_data_df
```

```python
#Drop the outlier circuit 9
expr_data_df = expr_data_df.drop([9])
expr_data_df
```

```python
#Pearson correlation coefficient
expr_data_df['Expr'].corr(expr_data_df['Ediff'])
```

```python
pylab.scatter(expr_data_df['Expr'], expr_data_df['Ediff'])
pylab.xlabel('Expressibility')
pylab.ylabel('Energy difference')
pylab.show()
```
